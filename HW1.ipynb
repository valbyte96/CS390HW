```python
'''dataReader.py
@author: Ted (Valerie) McCulloch
@date: 9/21/17
Reads a csv and writes two new csv files
with two separate data replacement methods
method 1: delete rows with missing data
method 2: insert the means where missing data exists

Note: code is definitely not optimized but gets the job done
Moreover, my display could be more sophisticated. I explored the data for
awhile and really was stumped on how to develop a meaningful representation of
all of it. Looking forward to discussing it in class. 
'''
import csv
import numpy as np
import matplotlib.pyplot as plt
#global lists
L1 = [] # data with missing rows deleted
L2 = [] # data with missing data replaced by means
L3 = [] # original data with no missing 
means1 = {}
means2 = {}
means3 = {}

'''opens csv, detects rows with missing data, and creates nested
lists for displaying data'''
def cleanData1():   
    with open('data.csv') as csvDataFile:
        # initiate reader and writers
        csvReader = csv.reader(csvDataFile)
        next(csvDataFile)
        for row in csvReader:
            empty = False
            for i in range(len(row)):
                if row[i] == '':
                    empty = True
                    break
            if empty == False:
                L1.append(row)
            L2.append(row)

    with open('Wholesale customers data.csv') as csvDataFile:
        # initiate reader and writers
        csvReader = csv.reader(csvDataFile)
        next(csvDataFile)
        for row in csvReader:
            L3.append(row)

    

'''computes all of the means'''
def computeMeans(L, dMean):
    nCol = len(L[0])
    for i in range(nCol):
        mean = 0 # init values
        col = 0
        for row in L:
            if row[i]=='':
                continue
            mean+=int(row[i])
            col+=1
        dMean[i]=int(mean/col)

'''Adds the means into the missing data'''                
def cleanData2(L, dMean):
    for row in L:
        for i in range(len(row)):
            if row[i] == '': # if empty
                row[i] = dMean[i]
    
                               
'''Displays data'''
def displayData():
    mList1 = tuple(means1.values())
    mList2 = tuple(means2.values())
    mList3 = tuple(means3.values())

    m1 = (mList1[2], mList1[3], mList1[4], mList1[5])
    m2 = (mList2[2], mList2[3], mList2[4], mList2[5])
    m3 = (mList3[2], mList3[3], mList3[4], mList3[5])

    nGroups = 4

    fig, ax = plt.subplots()
    index = np.arange(nGroups)
    bar_width = 0.3
    opacity = 0.8
    rects1 = plt.bar(index, m1, bar_width,
                     alpha=opacity,
                     color='b',
                     label='Deletion Method')     
    rects2 = plt.bar(index + bar_width, m2, bar_width,
                     alpha=opacity,
                     color='g',
                     label='Mean Replacement Method')
    rects2 = plt.bar(index + 2*bar_width, m3, bar_width,
                     alpha=opacity,
                     color='r',
                     label='Original')

     
    plt.xlabel('Methods')
    plt.ylabel('Value of Means')
    plt.title('Comparison of Means with Differing Data Cleaning Methods')
    plt.xticks(index + bar_width, ('Fresh', 'Milk', 'Grocery', 'Frozen'))
    plt.legend()
     
    plt.tight_layout()
    plt.show()




def main():
    cleanData1() # clean the data
    # compute means for each sort of data
    computeMeans(L1, means1)
    computeMeans(L2, means2)
    computeMeans(L3, means3)
    cleanData2(L2, means2) # insert new means into L2
    displayData()
    


main()
```
